```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
data_set = pd.read_excel('training_features.xls')
data_set
```

```python
# group CAD dollar by Zip Code to get the Target variable
Y = data_set.groupby('Zip Code')['CAD Dollar Amount'].sum()
print(len(Y))
```

```python
X = data_set.groupby('Zip Code').mean().iloc[:, -10:]
X.dropna(axis=1, how='all', inplace=True)
```

```python
model = sm.OLS(Y, X).fit()
```

```python
model.summary()
```
